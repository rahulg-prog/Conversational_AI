```python
import os
from dotenv import load_dotenv
from elevenlabs.client import ElevenLabs

load_dotenv()

client = ElevenLabs(
    api_key=os.getenv("ELEVENLABS_API_KEY")
)

response = client.conversational_ai.agents.get(agent_id="agent_1801ke6gspgzentb8rcbkxbkten5")
```

```python
response
```

```python
# Access agent details
print(f"Agent Name: {response.name}")
# To start a conversation with this agent
from elevenlabs.conversational_ai.conversation import Conversation
from elevenlabs.conversational_ai.default_audio_interface import DefaultAudioInterface

conversation = Conversation(
    client=client,
    agent_id=response.agent_id, # Using the ID from the response
    requires_auth=True,
    audio_interface=DefaultAudioInterface()
)
conversation.start_session()
```

```python
import signal
# Define how to handle the interrupt signal
signal.signal(signal.SIGINT, lambda sig, frame: conversation.end_session())
# Start your session
conversation.start_session()
```
